**1).** Сколько мужчин из города категории A?  

```python
len(data[(data['Gender'] == 'M') & (data['City_Category'] == 'A')])
```

```python
# уникальные ID можно посчитать так
len(np.unique(data[(data['Gender'] == 'M') & (data['City_Category'] == 'A')]['User_ID'].values))
```

**2).** Сколько женщин от 46 до 50, потративших (столбец Purchase) больше 20000 (условных единиц, в данном случае)?  

```python
len(data[(data['Gender'] == 'F') & (data['Age'] == '46-50') & (data['Purchase'] > 20000)])
```

**3).** Сколько NaN'ов в столбце Product_Category_3?

**4).** Какую долю (вещественное число от 0 до 1, округлить до 4-го знака) от всех покупателей составляют ВМЕСТЕ мужчины от 26 до 35 лет и женщины старше 36 лет (то есть нужно учесть несколько возрастных категорий)?

```python
men_26_35 = (data['Gender'] == 'M') & (data['Age'] == '26-35')
women_36plus = ((data['Gender'] == 'F') & (data['Age'] != '0-17') & (data['Age'] != '18-25') & (data['Age'] != '26-35'))
round(len(data[men_26_35 | women_36plus]) / len(data), 4)
```
